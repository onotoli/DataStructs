# **Структуры данных**

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
```

# Pushback с увеличением выделенной памяти в двое

```python
f = pd.read_csv("push_back_x2.csv", delimiter=";", names=['N', 'time'])
lnf = np.log(f)
```

```python
a = plt.subplot()
a.scatter(f['N'], f['time'], color=(1, 0 , 0))
plt.title("Увеличение выделенной памяти вдвое")
a.set_xlabel("Число элементов в массиве")
a.set_ylabel("Время обработки")
```

```python
def line(x, a, b):
    return a*x+b

popt, useless = curve_fit(line, lnf["N"].tolist(), lnf["time"].tolist())

a, b = popt

y = line(lnf['N'], a, b)
a, b
```

```python
a = plt.subplot()
a.scatter(lnf['N'], lnf['time'], color=(1, 0 , 0), alpha=1)
a.plot(lnf['N'], y, color=(0, 0, 1))
plt.title("Увеличение выделенной памяти вдвое")
a.set_xlabel("Число элементов в массиве")
a.set_ylabel("Время обработки")
```
